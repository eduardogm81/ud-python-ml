### Argumentos de read_csv
- sep: indica el separador, por defecto es la coma.
- Dtype: indica el tipo de la columna, None es el valor por defecto e indica que es pyhton quien debe de averiguar el tipo. Si queremos cargar columnas es dtype=["a"=np.float64, "b"=np.int32], donde np es NumPy
- header=none
- names: [] indicar el nombre de las columnas
- skiprows: Se salta las lineas
- skip_blank_lines(True)

### Carga de datos a través de la función open
open -> "r" sólo lectura, "rw" - lectura y escritura

```python
data3 = open(main_path + "/customer-churn-model/Customer Churn Model.txt", "r")
```

```python
data3 = open(main_path + "/customer-churn-model/Customer Churn Model.txt", "r")
cols = data3.readline().strip().split(",")
n_cols = len(cols)
```

```python
counter = 0
main_dict = {}
for col in cols:
    main_dict[col] = []
```

```python
for line in data3:
    values = line.strip().split(',')
    for i in range(len(cols)):
        main_dict[cols[i]].append(values[i])
    counter += 1

print ("El dataset tiene %d filas y %d columnas" % (counter, n_cols))
```

```python
df3 = pd.DataFrame(main_dict)
df3.head
```
